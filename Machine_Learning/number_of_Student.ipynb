```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
peopleData = pd.read_excel('number_of_student_kaohsiung.xlsx','people')  ## 抓取 number_of_student_kaohsiung.xlsx 檔案的 people 標籤
print(peopleData)
```

```python
peopleData.fillna(value=0)  ## 將空格填上 值為  0  的數值
```

```python
Eachyear_all_people = pd.DataFrame({'★設立別':['總計']})  ## 建立一個 DataFrame
print(Eachyear_all_people)
want_know_all = pd.merge(peopleData,Eachyear_all_people,on = ['★設立別'])  ## 將 peopleData 與 Eachyear_all_people中
                                                                           ## 合併 ★設立別(行) ，並顯示出有 總計(列) 的資訊 
                                                                           ## 兩筆資料的 行標籤  與  列標籤  也一併顯示
print(want_know_all)  ## 印出 want_know_all目前的資料
```

```python
year_elementry_all = want_know_all['國小']    ## 將 want_know_all裡面的「國小」(行) 中，所有的 列  資訊提取出來放入 year_elementry_all
print(year_elementry_all)                     ## 印出 year_elementry_all的資訊
```

```python
year_junior_all = want_know_all['國中']      ## 將 want_know_all裡面的「國中」(行) 中，所有的 列  資訊提取出來放入 year_junior_all
print(year_junior_all)                      ## 印出 year_junior_all的資訊
```

```python
## 高中職有分普通科、綜合高中、專業群、實用技能學程、進修部
## 本單元將上述五科的每學年各總計人數列出
senior_all = want_know_all.loc[:,['高級中等學校','高級中等學校.3','高級中等學校.6','高級中等學校.9','高級中等學校.12']]
##  將  want_know_all 的 標籤名稱中所有 列資訊 與  
##  行資訊中有「高級中等學校,高級中等學校.3,高級中等學校.6,高級中等學校.9,高級中等學校.12」的標籤資料
##  放入 senior_all
print(senior_all)     ## 印出 senior_all 的資訊

Each_senior_all = []  ## 設定 Each_senior_all 為 空陣列
for i in range(0,len(senior_all)):              ## 迴圈條件為 0 ~ senior_all陣列的長度(即有多少個 index)
    Each_senior_all.append(senior_all.iloc[i])  ## 將 index 的列資訊，放入 Each_senior_all陣列中


print(Each_senior_all)                          ## 印出 Each_senior_all 的資訊
```

```python
## 原始資料將高中職分為 普通科、綜合高中、專業群、實用技能學程、進修部
## 本單元將每學年的 普通科、綜合高中、專業群、實用技能學程、進修部 總計人數相加，得到每學年全台高中職就讀總人數
year_senior_all = []  ## 設定 year_senior_all 為空陣列
 
for i in range(0,len(Each_senior_all)):  ## 迴圈範圍為 0 ~ Each_senior_all 陣列的長度(即有多少個index)
    year_senior_all.append(np.sum(Each_senior_all[i]))  ## 將 Each_senior_all[i]內的所有數字相加(普通科、綜合高中、專業群、實用技能學程、進修部)
                                                        ## 再放入 year_senior_all 的陣列
print(year_senior_all)  ## 印出 year_senior_all ，即每一學年的高中職總人數
```

```python
##  原始資料將 大專校院分成  專科班、學士班
##  本單元將每學年的 專科班、學士班 總計人數相加，得到每學年全台高中職就讀總人數
university_all = want_know_all.loc[:,['大專校院','大專校院.3']]

##  將  want_know_all 的 標籤名稱中所有 列資訊 與  
##  行資訊中有「大專校院、大專校院.3」的標籤資料
##  放入 university_all

print(university_all)  ## 印出 university_all 的資訊

each_unviersitiy = []  ## 設定 each_unviersitiy 為 空陣列
for i in range(0,len(university_all)):                 ## 迴圈條件為 0 ~ university_all陣列的長度(即有多少個 index)
    each_unviersitiy.append(university_all.iloc[i])    ## 將 index 的列資訊，放入 each_unviersitiy 陣列中
    
print(each_unviersitiy) ## 印出 each_unviersitiy 的資訊
```

```python
year_university = []   ## year_university 為 空陣列
for i in range(0,len(each_unviersitiy)):                   ## 迴圈條件為 0 ~ each_unviersitiy 陣列的長度(即有多少個 index)
    year_university.append(np.sum(each_unviersitiy[i]))    ## 將 each_unviersitiy[i]內的所有數字相加(普通科、綜合高中、專業群、實用技能學程、進修部)
                                                           ## 再放入 year_university 的陣列
    
print(year_university) ## 印出 year_university 的資訊                          
               
```

```python
## 開始將資料視覺化

labelrange = 10**6    ## 設定 labelrange

plt.figure()          ## 定義一個圖像窗口
plt.xlabel('year')    ## 設定 X 座標軸標籤為 year
plt.xlim(90,110)      ## 設定 X 座標軸 範圍
plt.ylabel('number of student')  ## 設定 Y 座標軸標籤為 number of student
plt.ylim(year_elementry_all.min()-labelrange,year_elementry_all.max()+labelrange) 
## 設定 y 座標軸範圍 ，最小值為 「國小總計人數最小值 -labelrange」,最大值為「國小總計人數最大值 + labelrange」
plt.scatter(105 - year_elementry_all.index,year_elementry_all)
## 將每個點畫出來
## X： 105 - year_elementry_all的標籤 ，讓 X 值 從 97 到 105
## Y:  year_elementry_all國小總計人數
plt.show() ## 將圖畫出來
```
