```python
# Example : https://morvanzhou.github.io/tutorials/machine-learning/torch/3-02-classification/
#           https://discuss.pytorch.org/t/runtimeerror-multi-target-not-supported-newbie/10216/4

from torch.autograd import Variable
from IPython.display import display, clear_output
import torch
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.functional as activite
```

```python
# Step1. Build Dataset
np.random.seed(0)
number_data = np.ones( (100,2) , dtype = "float32" )

## about np.random.normal :  https://codertw.com/%E7%A8%8B%E5%BC%8F%E8%AA%9E%E8%A8%80/561625/
## choose data
raw_coordinate1 = np.random.normal(2 * number_data , 1)
raw_coordinate2 = np.random.normal(-2 * number_data , 1)

## define classification
target_class0 = np.zeros(100)
target_class1 = np.ones(100)

## merge data
raw_coordinate  = np.concatenate( ( raw_coordinate1 , raw_coordinate2 ) , 0 )
target_class  = np.concatenate( ( target_class0 , target_class1 ) , 0 )



## draw data
plt.title("Before classify")
plt.scatter(raw_coordinate[ : , 0] , raw_coordinate[ : , 1] )
plt.show()

plt.title("After classify")
plt.scatter(raw_coordinate[:, 0], raw_coordinate[:, 1], c=target_class , s=100, lw=0, cmap='RdYlGn')
plt.show()
```

```python
# Step2. build classifier Model

class classifier(torch.nn.Module):
    def __init__(self , number_feature , number_hidden , number_output):
        super(classifier ,self).__init__()
        self.hidden = torch.nn.Linear(number_feature,number_hidden)
        self.output = torch.nn.Linear(number_hidden,number_output)
        
    def forward(self , inputs):
        activite_value_1to2 = activite.relu(self.hidden(inputs))
        output_value = self.output(activite_value_1to2) 
        return output_value
    
net = classifier(number_feature = 2 , number_hidden = 10 , number_output = 2)
print(net)
```

```python
# Step3. build and train model
## information : http://cn.voidcc.com/question/p-kazqlthx-da.html
##               http://www.manongjc.com/article/35673.html
##               https://www.cnblogs.com/ranjiewen/p/10059490.html
##               https://medium.com/@chih.sheng.huang821/%E6%A9%9F%E5%99%A8-%E6%B7%B1%E5%BA%A6%E5%AD%B8%E7%BF%92-%E5%9F%BA%E7%A4%8E%E4%BB%8B%E7%B4%B9-%E6%90%8D%E5%A4%B1%E5%87%BD%E6%95%B8-loss-function-2dcac5ebb6cb

optimizer = torch.optim.SGD( net.parameters() , lr=0.002 )
loss_function = torch.nn.CrossEntropyLoss()
number_epochs = 100
inputs        = torch.from_numpy (raw_coordinate).type(torch.FloatTensor)
target_tensor = torch.from_numpy(target_class ).type(torch.LongTensor)

for epoch in range(number_epochs):
    output = net(inputs)
    loss = loss_function(output , target_tensor)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    ## visualization
    if ( epoch +1 ) % 2 == 0:
        print(" Train times : {} , error rate of modle  {} ".format( epoch + 1  , loss.data.numpy() ) ) 
        plt.cla()
        prediction = torch.max(activite.softmax(output),1)[1]
        target_prediction = prediction.data.numpy().squeeze()
        plt.scatter(raw_coordinate[:, 0], raw_coordinate[:, 1], c=target_prediction , s=100, lw=0, cmap='RdYlGn')
        accuracy = sum(target_class == target_prediction) / 200
        plt.text(1.5, -4, 'Accuracy=%.2f' % accuracy, fontdict={'size': 20, 'color':  'red'})
        #clear_output(wait = True)
        plt.pause(0.1)
```
